### Génération artificielle de données
1. Fonction rand_tri_gauss

Cette fonction génère des points dans trois classes différentes, où chaque classe suit une distribution gaussienne (ou normale). Les paramètres de la fonction définissent le nombre de points à générer et la distribution (moyenne et variance) de chaque gaussienne.

Elle renvoie un tableau de données où chaque point est représenté par ses coordonnées 2D (x, y), et une dernière colonne indiquant la classe à laquelle le point appartient (0, 1 ou 2).
Cette dernière colonne correspond aux étiquettes de classe associées aux points générés, c’est-à-dire 0, 1 ou 2.

2. Fonction rand_clown

La fonction rand_clown génère des données avec une distribution en forme de "visage de clown". Les points forment deux classes principales : l’une pourrait représenter un "visage", et l’autre des "yeux" ou des "traits". Cette fonction est principalement utilisée pour générer des données non linéaires et montrer des exemples de problèmes où une frontière linéaire ne suffit pas.

Comme rand_tri_gauss, elle renvoie un tableau avec les coordonnées (x, y) des points, et une dernière colonne indiquant à quelle classe (généralement 0 ou 1) appartient chaque point.
La dernière colonne représente les étiquettes de classe (0 ou 1) des points, en fonction de leur appartenance à l’une des deux régions principales du "clown".

3. Fonction rand_checkers

La fonction rand_checkers génère des données en forme de damier (ou "checkerboard"), souvent utilisées pour illustrer des exemples de classification complexe, où les classes sont alternées dans un motif en damier. Ici, les points de chaque classe sont intercalés, formant un motif répétitif.

Comme les autres fonctions, elle retourne un tableau de données avec les coordonnées (x, y) des points, et une colonne pour les classes (souvent 0 ou 1).
La dernière colonne contient les étiquettes des points, 0 ou 1, en fonction de la case (noire ou blanche) du damier à laquelle ils appartiennent.

#### Affichage de quelques jeux de données

```python
def plot_2d(X, y, xlabel,ylabel, title="Dataset"):
    # Tracer les points pour chaque classe avec des couleurs différentes
    for class_value in np.unique(y):
      # y == class_value me donne un tableau numpy booleen
        plt.scatter(X[y == class_value, 0], X[y == class_value, 1], label=f'Classe {int(class_value)}')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()
```

Avec rand_tri_gauss

```python
X_1,y_1 = tp.rand_tri_gauss()
plt.show()
plt.close("all")
plt.ion()
plt.figure(figsize=(19, 5))
plt.subplot(141)
plot_2d(X_1, y_1, xlabel="x_1", ylabel="x_2", title="Dataset 1")
```

Avec rand_clown

```python
X_2, y_2 = tp.rand_clown()
plt.figure(figsize=(19, 5))
plt.subplot(142)
plot_2d(X_2, y_2, xlabel="x_1", ylabel="x_2", title="Dataset 2")
```

Avec rand_checker

```python
X_3, y_3 = tp.rand_checkers()
plt.figure(figsize=(19, 5))
plt.subplot(143)
plot_2d(X_3, y_3, xlabel="x_1", ylabel="x_2", title="Dataset 3")
```
